Summary about data from csv file and parquet file:
1. csv: feature "VendorID" has missing values, parquet: feature "Vendor"ID has no missing values
2. csv: feature "payment_type" has missing valies, parquet: feature "payment_type" has no missiong values or missing values were filled by value `0`
3. one of Pandas behavior is if any numerical features have missing values the data type wil to be `float`
4. csv: the datetime type feature will be read as `object` or string type, parquet: the datetime type feature will be read as `datetime`

For learning purposes we will use schema from parquet file to get `int` type from `VendorID` and `payment_type` features. The date features for csv can be converted from string to datetime.

```python
csv_df['tpep_dropoff_datetime'] = pd.to_datetime(csv_df['tpep_dropoff_datetime'])
csv_df['tpep_pickup_datetime'] = pd.to_datetime(csv_df['tpep_pickup_datetime'])
```

```python
df_iter  = pd.read_csv("yellow_tripdata_2021-01.csv.gz", iterator=True, chunksize=100000)

while True:
    t_start = time()

    chunk_df = next(df_iter)

    chunk_df["tpep_pickup_datetime"] = pd.to_datetime(chunk_df["tpep_pickup_datetime"])
    chunk_df["tpep_dropoff_datetime"] = pd.to_datetime(chunk_df["tpep_dropoff_datetime"])

    chunk_df.to_sql(name="yellow_taxi_data", con=engine, if_exists="append")

    t_end = time()

    print(f"inserted another chunk, took {t_end-t_start} seconds")
```
